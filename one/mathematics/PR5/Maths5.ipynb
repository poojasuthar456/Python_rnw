Importing Libraries

```python
import numpy as np
import pandas as pd
```

PART - A: Vectors, norms, dot/cross/projection

```python
data = np.array([
    [80, 75, 85],
    [75, 90, 85],
    [70, 60, 80]
])
subjects = ["Maths", "Physics", "English"]
students = ["A", "B", "C"]
d1 = pd.DataFrame(data, columns=subjects, index=students)
d1 
```

```python
stu1 = data[0]
stu2 = data[1]

print("Student(A)-Marks:",stu1)
print("Student(B)-Marks:",stu2)
```

```python
# L1 Norm
norm1 = np.sum(np.abs(stu1))

# L2 Norm
norm2 = np.sqrt(np.sum(stu1 ** 2))

print("Using norm1:",norm1)
print("Using norm2:",norm2)
```

```python
from math import acos,degrees
dot_product = np.dot(stu1, stu2)

# Angle between vectors
cos_theta = dot_product / (np.linalg.norm(stu1) * np.linalg.norm(stu2))
angle = degrees(acos(cos_theta))

print("Dot Product Value:",dot_product)
print("Angle Between Vectors:",angle)
```

```python

v1 = stu1[:3]
v2 = stu2[:3]

cross = np.cross(v1, v2)
print("Cross product Value:",cross)
```

```python
from math import acos,degrees
projection = (np.dot(stu1, stu2) / np.dot(stu2, stu2)) * stu2
print("Projection on Vector:",projection)
```

PART - B: Matrix ops, transpose, det

```python
matrix = data
matrix
```

```python
# Addition
add = matrix + matrix
```

```python
# Multiplication
mul = np.dot(matrix, matrix.T)

print("Matrix Addition:",add)
print("Matrix Miltiplication:",mul)
```

```python
transpose = matrix.T
transpose
```

```python
# Determinant (square matrix required)
sq_matrix = matrix[:4, :4]
det = np.linalg.det(sq_matrix)
print("Determinant value of this Matrix:",det)
```

```python
# Inverse
inverse = np.linalg.inv(sq_matrix)
inverse
```

PART - D: Covariance eigenvalues/eigenvectors

```python
cov_matrix = np.cov(matrix.T)
cov_matrix
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigen Value:",eigenvalues)
print("Eigen Vector:",eigenvectors)
```

```python
from scipy.linalg import lu

P, L, U = lu(sq_matrix)
P, L, U
```

```python
U, S, Vt = np.linalg.svd(matrix)

U, S, Vt
```
