```python
import numpy as np

n = 1000  # number of tosses

# Option 1: use 0/1, where 1 = Head, 0 = Tail
flips = np.random.randint(0, 2, size=n)   # 0 or 1 with equal probability

heads_count = np.sum(flips == 1)
tails_count = np.sum(flips == 0)

heads_prob = heads_count / n
tails_prob = tails_count / n

print("Heads count:", heads_count, "Probability:", heads_prob)
print("Tails count:", tails_count, "Probability:", tails_prob)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data
np.random.seed(42)  # For reproducible results
coin_flips = np.random.randint(0, 2, size=1000)  # 0=Tails, 1=Heads
die_rolls = np.random.randint(1, 7, size=1000)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Coin toss: countplot
sns.countplot(x=coin_flips, ax=ax1)
ax1.set_title('Coin Toss Frequency (1000 flips)')
ax1.set_xlabel('0=Tails, 1=Heads')
ax1.set_ylabel('Count')

# Die rolls: countplot
sns.countplot(x=die_rolls, ax=ax2)
ax2.set_title('Die Roll Frequency (1000 rolls)')
ax2.set_xlabel('Face Value')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```
