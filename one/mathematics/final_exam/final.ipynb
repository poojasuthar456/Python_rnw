Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = pd.read_csv(r"D:\Python_rnw\one\mathematics\final_exam\employee_performance.csv")
```

Step 1: Central Tendency & Dispersion

Mean, Median, Mode of Salary 

```python
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
mode_salary = df['Salary'].mode()[0]

print("""Mean-:""",mean_salary,
      """, Median-:""",median_salary,
      """, Mode-:""",mode_salary)
```

Variance & Standard Deviation

```python
variance_projects = df['Projects_Completed'].var()
std_projects = df['Projects_Completed'].std()

print("""Variance:-""",variance_projects, """,Standard Deviaiton:-""", std_projects)
```

Step 2: Probability & Events

Probability of Promotion

```python
promotion_prob = (df['Promotion_Status'] == 'Yes').mean()

print("Probability of Promotion :-",promotion_prob)
```

Contingency Table 

```python
pd.crosstab(df['Department'], df['Promotion_Status'])
```

Conditional Probability

```python
high_perf = df[df['Performance_Score'] > 80]
conditional_prob = (high_perf['Promotion_Status'] == 'Yes').mean()

print("Condition Probability:-",conditional_prob)
```

Step 3: Distributions & Visualization

Histogram with Gaussian Curve

```python
plt.hist(df['Performance_Score'], bins=30, density=True)
mu, std = df['Performance_Score'].mean(), df['Performance_Score'].std()
x = np.linspace(df['Performance_Score'].min(), df['Performance_Score'].max(), 100)
plt.plot(x, stats.norm.pdf(x, mu, std))
plt.xlabel("Performance Score")
plt.ylabel("Density")
plt.show()
```

Skewness & Kurtosis of Salary

```python
skewness = df['Salary'].skew()
kurtosis = df['Salary'].kurtosis()

print(f"Skewness- {skewness:.2f}, Kurtosis- {kurtosis:.2f}")
```

Q-Q Plot

```python
stats.probplot(df['Projects_Completed'], plot=plt)
plt.show()
```

Step 4: Linear Algebra Application

Take First 5 Employees

```python
vectors = df[['Projects_Completed', 'Working_Hours']].head(5).values
```

Dot Product

```python
np.dot(vectors[0], vectors[1])
```

Norms

```python
n1 = np.linalg.norm(vectors[0])
n2 = np.linalg.norm(vectors[1])
print("Norm1:-",n1,"Norm2:-", n2)
```

Angle Between Two Employees

```python
cos_theta = np.dot(vectors[0], vectors[1]) / (n1 * n2)
angle = np.degrees(np.arccos(cos_theta))
print("Angel between two employee is:-",angle)
```
