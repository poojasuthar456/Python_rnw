CONFIDENCE INTERVAL

```python
age = data['age']

mean_age = age.mean()
std_age = age.std(ddof=1)
n = len(age)

confidence_level = 0.95
z = stats.norm.ppf(1 - (1 - confidence_level) / 2)

margin = z * (std_age / np.sqrt(n))

ci_lower = mean_age - margin
ci_upper = mean_age + margin

print("Mean Age is :",mean_age)
print("Lower Limit of Confidence Interval:",ci_lower)
print("Upper Limit of Confidence Interval:",ci_upper)
```

Z-TEST

```python
n = len(age)
sample_mean = np.mean(age)
sample_std = np.std(age, ddof=1)

mu = 50 

z_score = (sample_mean - mu) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean is:",sample_mean)
print("Sample standard Deviation is:",sample_std)
print("Z-score Value is:",z_score)
print("P-value is:",p_value)

alpha = 0.05

if p_value < alpha:
    result_ztest = "Reject H0"
else:
    result_ztest = "Accept H0"

result_ztest
```

T-TEST

```python
blood_pressure = data['blood_pressure']
n = len(blood_pressure)

mean = 50
alpha = 0.05

# One-sample t-test
t_statistic, p_value = stats.ttest_1samp(blood_pressure, mean)

print("Sample Mean:", np.mean(blood_pressure))
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject H0")
else:
    print("Accept H0")
```

CHI-SQUARE TEST

```python
from scipy.stats import chi2_contingency

Table = pd.crosstab(data['smoking_status'], data['diabetes'], margins=True, margins_name="Total")

print("Contingecy Table:\n", Table)
chi2, p, d_f, expected = chi2_contingency(Table)
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", d_f)

print("p-value:", p)
alpha = 0.05
if p < alpha:
    print("Reject H0: Smoking and diabetes are significantly associated.")
else:
    print("Fail to Reject H0: No significant association found.")
```

ANOVA TEST

```python
from scipy import stats
import pandas as pd

data=pd.read_csv(r'D:\Python_rnw\one\mathematics\PR-3\health_dataset_derivable_judgement.csv')

group_a = data[data["gender"] == "Male"]["bmi"]
group_b = data[data["gender"] == "Female"]["bmi"]

f_stat, p_anova = stats.f_oneway(group_a, group_b)

print("F-Value is:", f_stat)
print("P-Value is:", p_anova)

if p_anova < 0.05:
    anova_result = "Reject H0"
else:
    anova_result = "Accept H0"

anova_result
```
